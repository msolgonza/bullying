Estudio de los valores que toma cada columna:

```python
diccionario_tabla = {"Columna":[],"Valores":[]}

for col in cols:
    diccionario_tabla["Columna"].append(col)
    diccionario_tabla["Valores"].append(bullying[col].unique())
    
dataframe_tabla = pd.DataFrame(diccionario_tabla)
print(dataframe_tabla)
```
