En nuestro dataset hay muchas columnas (309) y casi todas duplicadas, ya que para cada respuesta obtenida, al string ya fue asignada una variable categórica antes de nuestro análisis.

Eliminamos las columnas que tengan variables numéricas a partir de strings, y nos quedamos con la variable string.

Ejemplo: de las columnas q1 y texto_q1 nos quedamos solo con texto_q1 y eliminamos q1

```python
import re

columnas_a_eliminar=[]
for col in cols:
    match = re.match(r'^texto_q(\d+)$', col)
    if match:
        numero = match.group(1)
        q_col = f'{numero}'
        if q_col in cols:
            columnas_a_eliminar.append(q_col)

# Eliminar las columnas identificadas
bullying.drop(columns=columnas_a_eliminar, inplace=True)

print(columnas_a_eliminar)
```
